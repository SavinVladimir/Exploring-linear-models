# Разведочный анализ данных

Разведывательный анализ данных предназначен для изучения основных свойств данных и нахождение закономерностей, аномалий и распределений.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("darkgrid")
```

```python
data = pd.read_csv('../data/Housing_Prices.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
#sns.boxplot(x=data["Avg. Area Income"])
#sns.boxplot(x=data["Avg. Area House Age"])
sns.boxplot(x=data["Avg. Area Number of Rooms"])
```

```python
sns.pairplot(data, diag_kind="kde")
```

```python
fig = plt.figure(figsize=(15, 12), dpi=200)

ax_1 = fig.add_subplot(3, 2, 1)
ax_1 = data['Avg. Area Income'].hist(label='Avg. Area Income')

ax_2 = fig.add_subplot(3, 2, 2)
ax_2 = data['Avg. Area House Age'].hist(label='Avg. Area House Age')

ax_3 = fig.add_subplot(3, 2, 3)
ax_3 = data['Avg. Area Number of Rooms'].hist(label='Avg. Area Number of Rooms')

ax_4 = fig.add_subplot(3, 2, 4)
ax_4 = data['Avg. Area Number of Bedrooms'].hist(label='Avg. Area Number of Bedrooms')

ax_5 = fig.add_subplot(3, 2, 5)
ax_5 = data['Area Population'].hist(label='Area Population')

ax_6 = fig.add_subplot(3, 2, 6)
ax_6 = data['Price'].hist(label='Price')

ax_1.legend()
ax_2.legend()
ax_3.legend()
ax_4.legend()
ax_5.legend()
ax_6.legend()
```

```python
sns.set_style("whitegrid")
mask = np.triu(np.ones_like(data.corr(), dtype=np.bool))
fig, ax = plt.subplots()
heatmap = sns.heatmap(data.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG', linewidths=1)
fig.set_size_inches(11.7, 8.27)
```
